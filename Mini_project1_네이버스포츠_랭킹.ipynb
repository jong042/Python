```python
import pandas as pd
import numpy as np
from datetime import date, timedelta
import requests
from bs4 import BeautifulSoup as bs
```

```python
d = date.today().strftime("%Y%m%d")
url = f"https://sports.news.naver.com/ranking/list.nhn?type=popular&date={d}"

response = requests.get(url, headers={'user-agent' : 'Mozilla/5.0'})
response
```

```python
sports = response.json()
sports
```

```python
news_list = sports["rankingNewsList"]["list"]
news_list
```

```python
df = pd.DataFrame(news_list)
df
```

```python
news_df = df[['rank', 'date', 'section', 'officeName', 'title', 'totalCount']]
news_df
```

```python
today = date.today().strftime("%Y-%m-%d")
file_name = f"sport-{today}_news.csv"
news_df.to_csv(file_name, index=False)
```
